```python
%%time

df = gse.pivot_samples('VALUE')
```

```python
info = gse.gpls['GPL2895'].table
info = info.set_index('ID')
```

```python
GB_LIST_DICT = info.GB_LIST[~info.GB_LIST.isna()].to_dict()
```

```python
df.index =df.index.map(GB_LIST_DICT)
```

```python
indice = df.index[~df.index.isna()]
```

```python
df = df.loc[indice]
df.head()
```

```python
df.index
```

---
### df duplicated() 처리한 후에 mygene 이용해서 gene mapping하기

```python
import mygene
```

```python
dup_df = df.loc[df[df.index.duplicated()].index]  ## 어떤 row를 살릴지 데이터 살펴보고 결정.
len(dup_df.index.unique()), dup_df.shape
```

```python
df_nodup = df.groupby(df.index).mean()
df_nodup.shape
```

```python
mg = mygene.MyGeneInfo()
mg_dict = mg.querymany(df_nodup.index,scopes='refseq,symbol,entrezgene,reporter,uniprot', species='human')
```

```python
mg_nodup = [ q_dict for q_dict in mg_dict if 'notfound' not in q_dict.keys()]
print(len(mg_nodup))
mg_nodup_df = pd.DataFrame.from_dict(mg_nodup).set_index('query')
mg_nodup_df.head()
```

```python
indice = mg_nodup_df.groupby('entrezgene')['_score'].idxmax()
```

```python
mg_nodup_df.loc[indice].head()
```

```python
dict_df = mg_nodup_df.loc[indice]
dict_entrezgene = dict_df.entrezgene.to_dict()
indice_n = df_nodup.index[~df_nodup.index.map(dict_entrezgene).isna()]
new_index = df_nodup.loc[indice_n].index.map(dict_entrezgene)
df_result = df_nodup.loc[indice_n]
df_result.index=new_index
df_result = df_result.fillna(0)
#df_result.to_csv('GSE83452_nodup.entrezid.txt',sep='\t')
```

### unigene

```python
import pandas_explode 
pandas_explode.patch()
```

```python
mg = mygene.MyGeneInfo()
mg_dict2 = mg.querymany(df_result.index,scopes='refseq,symbol,entrezgene,reporter,unigene', field=['entrezgene','unigene'], species='human')
mg_df2 = pd.DataFrame.from_dict(mg_dict2)
```

```python
mg_df2 = pd.DataFrame.from_dict(mg_dict2)
mg_df2 = mg_df2[['query','unigene','entrezgene']]
mg_df2 = mg_df2[~mg_df2.unigene.isna()]
```

```python
mg_df2.unigene = [ [m] if type(m)==str else m for m in mg_df2.unigene ]
```

```python
mg_df2.head()
```

```python
mg_df2.explode('unigene')[['entrezgene','unigene']].to_csv('GSE24807.unigene.csv', index=False)
```
