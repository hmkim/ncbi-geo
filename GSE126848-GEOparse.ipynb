```python
gse = GEOparse.get_GEO("GSE126848")
```

```python
gse.gsms['GSM3615293'].metadata
```

```python
gsm_id_dict=dict()
for gsm_id in gse.gsms:
    desc = gse.gsms[gsm_id].metadata['description']
    gsm_id_dict[desc[0].zfill(4)] = gsm_id
#    print(desc[0].zfill(4) ,gsm_id)
```

```python
gse.metadata['supplementary_file']
```

```python
df = pd.read_csv(gse.metadata['supplementary_file'][0], sep='\t', index_col=0)
```

```python
df.head()
```

```python
df.columns = df.columns.map(str).map(gsm_id_dict)
```

```python
df.head()
```

```python
import mygene
mg = mygene.MyGeneInfo()
mg_df = mg.getgenes(df.index, fields='entrezgene',as_dataframe=True)
df.insert(0,'entrezgene',mg_df.entrezgene)

import numpy as np
df.entrezgene = df.entrezgene.astype(str)
df_nodup=df[df.entrezgene!='nan']
```

```python
dup_entrez=df_nodup[df_nodup.entrezgene.duplicated()].entrezgene.tolist()
sub_df = df[df.entrezgene.isin(dup_entrez)]
sub_df = sub_df.reset_index()
sub_df
```

```python
# sub_df['SUM']=sub_df.iloc[:,1:].sum(axis=1)
# df_nodup = df_nodup.drop(sub_df[sub_df['SUM']==0].index)
# df_nodup.shape
```

```python
sub_df.groupby(sub_df.entrezgene).min().key
```

```python
df_nodup2 = df[df.entrezgene!='nan']
df_nodup2 = df_nodup2[~df_nodup2.index.isin(sub_df.groupby(sub_df.entrezgene).min().key)]
df_nodup2.shape
```

```python
df_nodup2[df_nodup2.entrezgene.duplicated()]
```

```python
df_nodup2 = df_nodup2.set_index('entrezgene',drop=True)
df_nodup2.head()
```

```python
df_nodup2.to_csv('GSE126848_nodup.entrezid.txt',sep='\t')
```

### unigene

```python
import pandas_explode 
pandas_explode.patch()

df_result = df_nodup2
mg = mygene.MyGeneInfo()
mg_dict2 = mg.querymany(df_result.index,scopes='refseq,symbol,entrezgene,reporter,unigene', field=['entrezgene','unigene'], species='human')
mg_df2 = pd.DataFrame.from_dict(mg_dict2)
```

```python
mg_df2 = pd.DataFrame.from_dict(mg_dict2)
mg_df2 = mg_df2[['query','unigene','entrezgene']]
mg_df2 = mg_df2[~mg_df2.unigene.isna()]
mg_df2.unigene = [ [m] if type(m)==str else m for m in mg_df2.unigene ]
mg_df2.head()
```

```python
mg_df2.explode('unigene')[['entrezgene','unigene']].to_csv('GSE126848.unigene.csv', index=False)
```
